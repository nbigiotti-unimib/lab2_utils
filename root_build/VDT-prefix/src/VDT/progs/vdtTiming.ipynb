```python
import imp
import timeit
import numpy as np

import vdt
from vdt import *
print dir(vdt)

def norf(vi):
    requires = ['CONTIGUOUS', 'ALIGNED']
    vi = np.asanyarray(vi)
    vi = np.require(vi, np.single, requires)
    return vi

def nord(vi):
    requires = ['CONTIGUOUS', 'ALIGNED']
    vi = np.asanyarray(vi)
    vi = np.require(vi, np.double, requires)
    return vi


print vdt_arch()
print vdt_sin(nord([1,2,3,4]))
print vdt_sin(norf([1,2,3,4]))
print vdt_atan2(norf([1,2,-3,-4]),norf([1,-2,3,-4]))
print vdt_sincos(norf([1,2,3,4]))
print vdt_exp(norf([1,2,3,4]))


xx = nord(np.linspace(-np.pi, np.pi, 2001))
xf = norf(np.linspace(-np.pi, np.pi, 2001))
yf = norf(np.linspace(-np.pi, np.pi, 2001))
zf = norf(np.linspace(-np.pi, np.pi, 2001))

xl = norf(np.linspace(0., 2.*np.pi, 20001))

xo = numpy.empty_like(xf)

def nid():
    global xf,yf,zf
    return np.copy(xf)

def nfma():
    global xf,yf,zf
    return xf*yf+zf

def nsqrt():
    global xf,yf,zf
    return  np.sqrt(np.abs(xf))

def nsqrtl():
    global xl
    return  np.sqrt(xl)



def nsc() :
    global xx
    s = np.sin(xx)
    c = np.cos(xx)
    return (s,c)

def nscf() :
    global xf
    s = np.sin(xf)
    c = np.cos(xf)
    return (s,c)



def vid():
    global xf,yf,zf
    return  vdt_identity(xf)

def vidl():
    global xl
    return  vdt_identity(xl)

def vfma():
    global xf,yf,zf
    return  vdt_fma(xf,yf,zf)

def vsqrt():
    global xf,yf,zf
    return  vdt_sqrt(np.abs(xf))

def vsqrtl():
    global xl
    return  vdt_sqrt(xl)



def vsc() :
    global xx
    return vdt_sincos(xx)

def vscf() :
    global xf
    return vdt_sincos(xf)


def nex() :
    global xx
    return np.exp(xx)

def nexf() :
    global xf
    return np.exp(xf)


def vex() :
    global xx
    return vdt_exp(xx)

def vexf() :
    global xf
    return vdt_exp(xf)


#wake up avx
bha = timeit.timeit("vfma()", setup="from __main__ import vfma",number=100000)
bha = timeit.timeit("vsqrtl()", setup="from __main__ import vsqrtl",number=100000)


print "timing identity"
print(timeit.timeit("nid()", setup="from __main__ import nid",number=100000))
print(timeit.timeit("vid()", setup="from __main__ import vid",number=100000))
print(timeit.timeit("vidl()", setup="from __main__ import vidl",number=100000))

print "timing fma"
print(timeit.timeit("nfma()", setup="from __main__ import nfma",number=100000))
print(timeit.timeit("vfma()", setup="from __main__ import vfma",number=100000))


print "timing sqrt"
print(timeit.timeit("nsqrt()", setup="from __main__ import nsqrt",number=100000))
print(timeit.timeit("nsqrtl()", setup="from __main__ import nsqrtl",number=100000))
print(timeit.timeit("vsqrt()", setup="from __main__ import vsqrt",number=100000))
print(timeit.timeit("vsqrtl()", setup="from __main__ import vsqrtl",number=100000))


print "timing exp"
print(timeit.timeit("nex()", setup="from __main__ import nex",number=100000))
print(timeit.timeit("nexf()", setup="from __main__ import nexf",number=100000))
print(timeit.timeit("vex()", setup="from __main__ import vex",number=100000))
print(timeit.timeit("vexf()", setup="from __main__ import vexf",number=100000))


print "timing sincos"
print(timeit.timeit("nsc()", setup="from __main__ import nsc",number=100000))
print(timeit.timeit("nscf()", setup="from __main__ import nscf",number=100000))
print(timeit.timeit("vsc()", setup="from __main__ import vsc",number=100000))
print(timeit.timeit("vscf()", setup="from __main__ import vscf",number=100000))

```
